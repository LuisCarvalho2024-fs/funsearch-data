```python
from collections.abc import Sequence
from typing import Any

from CodeManipulator import code_utils
from Sampler import Sampler
from Evaluator import Evaluator
from Database.ProgramsDatabase import ProgramsDatabase
def _extract_function_names(specification: str) -> tuple[str, str]:
  """Returns the name of the function to evolve and of the function to run."""
  run_functions = list(
      code_utils.yield_decorated(specification, 'funsearch', 'run'))
  if len(run_functions) != 1:
    raise ValueError('Expected 1 function decorated with `@funsearch.run`.')
  evolve_functions = list(
      code_utils.yield_decorated(specification, 'funsearch', 'evolve'))
  if len(evolve_functions) != 1:
    raise ValueError('Expected 1 function decorated with `@funsearch.evolve`.')
  return evolve_functions[0], run_functions[0]
```

```python
os.environ["FUNSEARCH_DEBUG"] = "False"
with open('problem3.py', 'r') as file:
    specification = file.read() 

"""Launches a FunSearch experiment."""
function_to_evolve, function_to_run = _extract_function_names(specification)

template = code_utils.text_to_program(specification)
database = ProgramsDatabase(function_to_evolve, template)
inputs = [0]

evaluator = Evaluator(database, template, function_to_evolve, function_to_run, inputs)

initial = template.get_function(function_to_evolve).body
evaluator.analyse(initial, island_id=None, version_generated=None)

sampler = Sampler(database, evaluator)
sampler.sample(500)

database.database_state()
sampler.experiment_state()
```
